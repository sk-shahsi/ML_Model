```python
# use difference value from n_estimators
n=[50,60,100,200,300,400,500,600,700,800,900,1000]
accuracy =[]
for i in n:
  model = RandomForestClassifier(n_estimators=i)
  model.fit(x_train,y_train)
  ans = model.predict(x_test)
  acc = accuracy_score(y_test,ans)
  accuracy.append(acc)
```

```python
accuracy
```

```python
# Grid Search CV
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators':[100,200,300],
    'max_depth':[None,5,10],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'criterion':['gini','entropy']
}

```

```python
model2 = RandomForestClassifier(random_state=0)
grid_search = GridSearchCV(estimator = model2,param_grid=param_grid,cv=2)
grid_search.fit(x_train,y_train)
```

```python
model3 = RandomForestClassifier(n_estimators=200,max_depth=10,min_samples_leaf=2,criterion='entropy')
model3.fit(x_train,y_train)
ans = model3.predict(x_test)
accuracy_score(y_test,ans)
```
